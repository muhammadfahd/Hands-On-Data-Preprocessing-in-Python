## Creating plots using pandas
Pandas makes plotting simpler and faster.
### Basic Pandas Plot
| Function                        | Purpose         | Example                                        |
| ------------------------------- | --------------- | ---------------------------------------------- |
| `df.plot()`                     | General plot    | `df['age'].plot()`                             |
| `.plot(kind='hist')`            | Histogram       | `df['age'].plot(kind='hist')`                  |
| `.plot(kind='box')`             | Boxplot         | `df['income'].plot(kind='box')`                |
| `.plot(kind='bar')`             | Bar chart       | `df['gender'].value_counts().plot(kind='bar')` |
| `.plot(kind='line')`            | Line plot       | `df.plot(kind='line')`                         |
| `.plot(kind='scatter', x=, y=)` | Scatter plot    | `df.plot(kind='scatter', x='age', y='income')` |
| `.plot(subplots=True)`          | Create subplots | `df[['age','income']].plot(subplots=True)`     |


```python
# creating histogram
d_Set['age'].plot(kind='hist')
```

```python
# creating scatter plot
d_Set.plot.scatter(x='hoursPerWeek', y='age')
```
